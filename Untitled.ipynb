## Real Estate EDA

Below we will query the US Realty API provided by APIdojo.net located at https://rapidapi.com/apidojo/api/realty-in-us/
I am limited to 200 results per call so I will create a for loop in order to alter the page offset by 199 and concatenate
the resultant response to a single dataframe. We are limited to 5 calls per second so I have introduced a time delay in order
to avoid exceeding our call limit. 

Once I have all the real estate data in a dataframe I will perform some basic cleaning and EDA. 

```python

import requests
df = pd.DataFrame()
url = "https://realty-in-us.p.rapidapi.com/properties/list-sold"
for i in range(0,9500,199): #The API is limited to an offset less than 10,000
    x = str(i)
    querystring = {"city":"Miami","state_code":"FL","offset":x,"limit":"200","sort":"relevance"}
    headers = {
        "X-RapidAPI-Key": "4b9e4d4e78mshe48c955a29f17a2p1dc9a1jsn959cf97c82a8",
        "X-RapidAPI-Host": "realty-in-us.p.rapidapi.com"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    json = response.json()
    frame = pd.json_normalize(json['listings'])
    new_frame = pd.concat([frame,df], ignore_index = True)
    df = new_frame
    time.sleep(.3)
df.info()
    
```

next I would like to clean up the lot_size field. lot_size contains two different measurements square feet and acres. I will standard the lot size by square footage and assign it to a new columng called lot_square_feet

```python
temp_df = df[df['lot_size'].notna()] # separate out NaN values to a separate df

# separate out acre values into a new column
df['acres'] = temp_df[temp_df['lot_size'].str.contains('ac')].loc[0:,'lot_size'].str.split(" ").str[0]
```

```python
df['acres'] = pd.to_numeric(df['acres'])
```

```python
temp_df = df[df['lot_size'].notna()] # separate out NaN values to a separate df

# separate out sqft values into a new column
df['sq1'] = temp_df[temp_df['lot_size'].str.contains('sq')].loc[0:,'lot_size'].str.split(" ").str[0].str.replace(",","")
df['sq1'] = pd.to_numeric(df['sq1'])
```

```python
# Create a lot square feet column that takes values from sq1 and acres, converting acres to square feet
df['lot_square_feet'] = (df[['acres', 'sq1']] * [43560, 1]).max(axis=1)
```
